```python
import numpy as np # calculations with arrays
import pandas as pd # user-friendly DataFrames for data representation
import sklearn # machine learning algorithms
from sklearn import ensemble, linear_model
from sklearn.metrics import log_loss

import matplotlib.pyplot as plt # import plot functions
# necessary to plot in jupyter notebook:
%matplotlib inline
import seaborn as sns # make plots beautiful
```

```python
train = pd.read_csv('competition_zero/train2.csv')
test = pd.read_csv('competition_zero/test2.csv')
sample_submission = pd.read_csv('competition_zero/sample_submission.csv')
```

```python
t1 = pd.get_dummies(train.loc[:101608, 'team1'])
t2 = pd.get_dummies(train.loc[:101608, 'team2'])
```

```python

t2.columns = t2.columns * -1
```

```python
t = pd.concat([t1, t2],axis=1)
t2.columns = t2.columns * -1
t1.columns = t1.columns * -1
```

```python
tr = pd.concat([t2, t1],axis=1)
t = pd.concat([t, tr],axis=0)
```

```python
t[:3]
```

```python
target1 = []
for i in range(101609):
    if train.target[i] == False:
        target1 += [True]
    else:
        target1 += [False]
```

```python
target1
```

```python
targ = []
target0 = train.target
for i in range(101609):
    targ += [target0[i]]
targ[:4]
```

```python
for i in range(101609):
    targ += [target1[i]]
len(targ)
```

```python
from sklearn import cross_validation
train_data, test_data, train_labels, test_labels = cross_validation.train_test_split(t, targ, 
                                                                                     test_size = 0.3)
```

### ЛЕС

```python
from sklearn.ensemble import RandomForestClassifier
forest_regressor = RandomForestClassifier(random_state = 0, max_depth = 20, n_estimators = 50)
```

```python
#score = cross_validation.cross_val_score(forest_regressor, train_data, train_labels, scoring = 'log_loss', cv = 3)
```

```python
forest_regressor.fit(train_data, train_labels)
```

```python
forest_prediction_both = forest_regressor.predict_proba(test_data)
```

```python
forest_prediction = []
for c in range(60966):
    forest_prediction += [forest_prediction_both[c][1]]
```

```python
log_loss(test_labels, forest_prediction)
```

## Линия

```python
liner_regressor = linear_model.LogisticRegression()
```

```python
liner_regressor.fit(train_data, train_labels)
```

```python
liner_prediction_both = liner_regressor.predict_proba(test_data)
```

```python
liner_prediction_both.shape
```

```python
liner_prediction = []
for c in range(60966):
    liner_prediction += [liner_prediction_both[c][1]]
```

```python
cross_validation.cross_val_score(liner_regressor, train_data, train_labels, scoring = 'log_loss', cv = 3)
```

```python
log_loss(test_labels, liner_prediction_both)
```

## Бустинг

```python
import xgboost as xgb
boost_regressor = xgb.XGBClassifier(learning_rate=0.1, max_depth=5, n_estimators=20, min_child_weight=3)
```

```python
boost_regressor.fit(train_data, train_labels)
```

```python
boost_prediction_both = boost_regressor.predict_proba(test_data)
```

```python
boost_prediction = []
for c in range(30483):
    boost_prediction += [boost_prediction_both[c][1]]
```

```python
cross_validation.cross_val_score(boost_regressor, train_data, train_labels, scoring = 'log_loss', cv = 3)
```

```python
log_loss(test_labels, boost_prediction)
```

## Композиция

```python
super_prediction = []
for i in range(30483):
    super_prediction += [liner_prediction[i] * 0.7 + boost_prediction[i] * 0.3]

```

```python
log_loss(test_labels, super_prediction)
```

```python
submission = sample_submission.copy()
submission.target = boost_regressor.predict_proba(test) # notice here that we can refer to a column 'target' in two ways
submission.to_csv('boost_submission.csv', index=False)
```
