```python
import numpy as np # calculations with arrays
import pandas as pd # user-friendly DataFrames for data representation
import sklearn # machine learning algorithms
from sklearn import ensemble, linear_model
from sklearn.metrics import log_loss

import matplotlib.pyplot as plt # import plot functions
# necessary to plot in jupyter notebook:
%matplotlib inline
import seaborn as sns # make plots beautiful
```

```python
train = pd.read_csv('competition_zero/train2.csv')
test = pd.read_csv('competition_zero/test2.csv')
sample_submission = pd.read_csv('competition_zero/sample_submission.csv')
```

```python
team2 =  pd.DataFrame( {'team_id' : train[:]['team2'], 'score' : train[:]['score1'],'win' : train[:]['target'] })
team1 =  pd.DataFrame( {'team_id' : train[:]['team1'], 'score' : train[:]['score1'],'win' : train[:]['target'] })
team1.sort(columns = 'team_id', inplace = True)
team1[:10]
team2.sort(columns = 'team_id', inplace = True)
```

```python
groped = team1.groupby('team_id')
groped2 = team2.groupby('team_id')

tmp = groped['score'].sum()
tmp1 = groped['win'].sum()
tmp3 =  team1.groupby('team_id').size()
tmp2 = team1.drop_duplicates(['team_id'])[:]['team_id'].values

qtmp = groped2['score'].sum()
qtmp1 = groped2['win'].sum()
qtmp2 = team2.drop_duplicates(['team_id'])[:]['team_id'].values
qtmp3 = team2.groupby('team_id').size()
tmp1  = (tmp1+qtmp1)/(tmp3+qtmp3)
qtmp1 = groped2['score'].sum() * 0


q2 = pd.DataFrame({'score_sum':qtmp.values, 'win': qtmp1.values, 'team_id': qtmp2})
q = pd.DataFrame({'score_sum':tmp.values, 'win': tmp1.values, 'team_id': tmp2})
```

```python
tee = q + q2
tee.describe()
tee['team_id'] = tee['team_id'] / 2
tee['team_id'].astype('int64')

tee[:10]
```

```python
from sklearn import cross_validation
train_data, test_data, train_labels, test_labels = cross_validation.train_test_split(train, train.target, 
                                                                                     test_size = 0.3)
```

```python
predict = []
k = tee[tee.team_id == 6]['win']
k
```

```python
for c  in test_data.index:
    r = tee[tee.team_id == test_data.team1[c]]['win'].values
    s = tee[tee.team_id == test_data.team2[c]]['win'].values
    w = float(r[0])/ float((s[0] + r[0]))
    predict += [w]
```
