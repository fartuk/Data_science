```python
def r_metric(x,y):
    tmp = 1.0 - ((x - y) ** 2.0).sum() / ((x - x.mean()) ** 2.0).sum()
    return np.sign(tmp) * np.sqrt(np.abs(tmp)) * 1.0
```
